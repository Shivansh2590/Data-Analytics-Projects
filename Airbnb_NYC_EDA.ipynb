# Performing Data Cleaning

## Checking Missing Values

```python
df.isnull().sum()
```

```python
# % of missing values per column
(df.isnull().sum() / df.shape[0]) * 100
```

```python
# Total missing % across dataset
(df.isnull().sum().sum()) / (df.shape[0] * df.shape[1]) * 100
```

## Handling Missing Values

```python
# 'name' and 'host_name' — fill with 'Unknown'
df['name'].fillna('Unknown', inplace=True)
df['host_name'].fillna('Unknown', inplace=True)

# 'last_review' — fill with 'No Review'
df['last_review'].fillna('No Review', inplace=True)

# 'reviews_per_month' — fill with 0 (no reviews)
df['reviews_per_month'].fillna(0, inplace=True)

print('Missing values handled ✅')
df.isnull().sum()
```

## Checking for Duplicates

```python
df.duplicated().sum()
```

## Handle Outliers in Price Column

```python
# Check price distribution
print('Max price:', df['price'].max())
print('Listings with price = 0:', (df['price'] == 0).sum())

# Remove zero-price listings and extreme outliers (>$1000)
df = df[(df['price'] > 0) & (df['price'] <= 1000)]
print(f'Cleaned dataset shape: {df.shape}')
```

```python
# Boxplot to visualize outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['price'], color='#FF5A5F')
plt.title('Price Distribution — Outlier View', fontsize=13, fontweight='bold')
plt.xlabel('Price per Night ($)')
plt.tight_layout()
plt.show()
```

# Performing Exploratory Data Analysis

## i) Univariate Analysis

```python
# --- Room Type Distribution ---
plt.figure(figsize=(7, 5))
room_counts = df['room_type'].value_counts()
colors = ['#FF5A5F', '#00A699', '#FC642D']
plt.pie(room_counts, labels=room_counts.index, autopct='%1.1f%%',
        colors=colors, startangle=90, wedgeprops={'edgecolor': 'white', 'linewidth': 2})
plt.title('Room Type Distribution — NYC Airbnb', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
print('Entire home/apartment is the most common listing type (~52%)')
```

```python
# --- Neighbourhood Group Distribution ---
plt.figure(figsize=(8, 5))
ng_counts = df['neighbourhood_group'].value_counts()
sns.barplot(x=ng_counts.index, y=ng_counts.values, palette='Set2')
plt.xlabel('Borough')
plt.ylabel('Number of Listings')
plt.title('Listings by NYC Borough', fontsize=13, fontweight='bold')
for i, v in enumerate(ng_counts.values):
    plt.text(i, v + 100, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
print('Manhattan and Brooklyn account for over 70% of all listings')
```

```python
# --- Price Distribution ---
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=60, color='#FF5A5F', edgecolor='white')
plt.xlabel('Price per Night ($)')
plt.title('Distribution of Airbnb Listing Prices', fontsize=13, fontweight='bold')
plt.axvline(df['price'].median(), color='black', linestyle='--',
            label=f"Median: ${df['price'].median():.0f}")
plt.legend()
plt.tight_layout()
plt.show()
print(f"Median price: ${df['price'].median():.0f}/night | Mean: ${df['price'].mean():.0f}/night")
```

```python
# --- Top 10 Most Listed Neighbourhoods ---
top_hoods = df['neighbourhood'].value_counts().head(10)
plt.figure(figsize=(10, 5))
bars = plt.barh(top_hoods.index[::-1], top_hoods.values[::-1], color='#00A699')
plt.xlabel('Number of Listings')
plt.title('Top 10 Neighbourhoods by Listing Count', fontsize=13, fontweight='bold')
for bar, val in zip(bars, top_hoods.values[::-1]):
    plt.text(bar.get_width() + 5, bar.get_y() + bar.get_height()/2, str(val), va='center')
plt.tight_layout()
plt.show()
```

## ii) Bivariate Analysis

```python
# --- Average Price by Borough ---
plt.figure(figsize=(8, 5))
avg_price = df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
sns.barplot(x=avg_price.index, y=avg_price.values, palette='Reds_r')
plt.ylabel('Average Price ($)')
plt.title('Average Price per Night by Borough', fontsize=13, fontweight='bold')
for i, v in enumerate(avg_price.values):
    plt.text(i, v + 1, f'${v:.0f}', ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
print('Manhattan has the highest average nightly price')
```

```python
# --- Price by Room Type ---
plt.figure(figsize=(8, 5))
sns.boxplot(x='room_type', y='price', data=df, palette=['#FF5A5F','#00A699','#FC642D'])
plt.title('Price Distribution by Room Type', fontsize=13, fontweight='bold')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.tight_layout()
plt.show()
print('Entire home/apartment listings are significantly more expensive')
```

```python
# --- Number of Reviews vs Price ---
plt.figure(figsize=(9, 5))
plt.scatter(df['number_of_reviews'], df['price'], alpha=0.3,
            c='#FF5A5F', s=10, edgecolors='none')
plt.xlabel('Number of Reviews')
plt.ylabel('Price ($)')
plt.title('Number of Reviews vs Listing Price', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
print('Listings with many reviews tend to have lower prices (better value perception)')
```

```python
# --- Availability vs Borough ---
plt.figure(figsize=(9, 5))
sns.boxplot(x='neighbourhood_group', y='availability_365', data=df, palette='Set2')
plt.title('Availability (365 days) by Borough', fontsize=13, fontweight='bold')
plt.xlabel('Borough')
plt.ylabel('Days Available per Year')
plt.tight_layout()
plt.show()
```

## iii) Multivariate Analysis

```python
# --- Average Price: Borough × Room Type ---
pivot_price = df.groupby(['neighbourhood_group', 'room_type'])['price'].mean().unstack()
pivot_price.plot(kind='bar', figsize=(10, 6), colormap='Set1', edgecolor='white', width=0.7)
plt.title('Average Price by Borough and Room Type', fontsize=13, fontweight='bold')
plt.xlabel('Borough')
plt.ylabel('Avg Price ($)')
plt.legend(title='Room Type')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# --- Listing Count: Borough × Room Type ---
pivot_count = df.groupby(['neighbourhood_group','room_type']).size().unstack(fill_value=0)
pivot_count.plot(kind='bar', figsize=(10, 6), colormap='Set2', edgecolor='white', width=0.7)
plt.title('Listing Count by Borough and Room Type', fontsize=13, fontweight='bold')
plt.xlabel('Borough')
plt.ylabel('Number of Listings')
plt.legend(title='Room Type')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# --- Geographic Distribution of Listings ---
plt.figure(figsize=(12, 9))
colors_map = {'Manhattan': '#FF5A5F', 'Brooklyn': '#00A699',
              'Queens': '#FC642D', 'Bronx': '#7B0D1E', 'Staten Island': '#484848'}
for grp, grp_data in df.groupby('neighbourhood_group'):
    plt.scatter(grp_data['longitude'], grp_data['latitude'],
                c=colors_map.get(grp, 'gray'), s=1, alpha=0.4, label=grp)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographic Distribution of NYC Airbnb Listings', fontsize=14, fontweight='bold')
plt.legend(title='Borough', markerscale=6)
plt.tight_layout()
plt.show()
```

```python
# --- Top 10 Hosts by Listing Count ---
top_hosts = df['host_name'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_hosts.values, y=top_hosts.index, palette='Reds_r')
plt.xlabel('Number of Listings')
plt.title('Top 10 Hosts by Number of Listings', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
print('A small number of professional hosts manage a large share of listings')
```

## Key Insights & Findings

### 1. Room Type
- **Entire home/apt (~52%)** is most common, followed by Private rooms (~46%).
- Shared rooms are rare (<2%) and cheapest.

### 2. Borough Trends
- **Manhattan & Brooklyn** hold 70%+ of all listings.
- **Manhattan** has the highest average price (~$196/night).
- **Bronx** is the most affordable borough.

### 3. Price Patterns
- Median price is **~$106/night**; distribution is right-skewed.
- Entire apartments cost 2–3× more than private rooms.

### 4. Reviews
- Listings with more reviews tend to be cheaper — high-volume budget stays.

### 5. Geography
- Listings cluster densely in **Lower Manhattan, Williamsburg, and Bushwick**.

## Conclusion
The NYC Airbnb dataset reveals strong geographic and room-type driven price differences. Manhattan dominates in price while Brooklyn leads in volume. Professional hosts with multiple listings are a notable market segment.
